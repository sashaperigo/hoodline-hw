```python
def get_locations(df):
    cities = set()
    for i, row in df.iterrows():
        city = row['location_city']
        region = row['location_region']
        country = row['location_country_code']
        
        if type(city) != str or type(region) != str or type(country) != str:
            continue

        cities.add("%s, %s, %s" % (city, region, country))
    return cities
```

```python
funding = pd.read_csv('crunchbase_data.csv')
cities = get_locations(funding)
```

## Get latitude and longitude

```python
def get_coordinates(cities):
    coordinate_map = {}
    failed = []
    for city in cities:
        try:
            location = geocoders.GeoNames(username=random.choice(keys)).geocode(city)
            if location == None:
                continue
            print location
            coordinates = (location.latitude, location.longitude)
            coordinate_map[city] = coordinates
        except:
            print sys.exc_info()[0]
            continue
    
    return coordinate_map
```

## Cluster location coordinates to calculate regions

```python
def distance(x, y):
    lat1 = x[0]
    lat2 = y[0]
    lon1 = x[1]
    lon2 = y[1]
    p = 0.017453292519943295 #Pi/180
    a = 0.5 - cos((lat2 - lat1) * p)/2 + cos(lat1 * p) * cos(lat2 * p) * (1 - cos((lon2 - lon1) * p)) / 2
    return 7918 * asin(sqrt(a)) 
```

```python
def find_regions(coordinate_map):
    done = set()
    clusters = []
    for city_name, city_coords in coordinate_map.items():
        if city_name in done:
            continue
        cluster = [city_name]
        for neighbor_name, neighbor_coords in coordinate_map.items():
            if city_name == neighbor_name:
                continue
            if neighbor_name in done:
                continue
            miles = distance(city_coords, neighbor_coords)
            if miles < 75:
                cluster.append(neighbor_name)
                done.add(neighbor_name)
        done.add(city_name)
        clusters.append(cluster)
    return clusters
```

```python
clusters = find_regions(x)
```

```python
def get_city_counts(df):
    city_counts = defaultdict(int)
    for i, row in df.iterrows():
        city = row['location_city']
        region = row['location_region']
        country = row['location_country_code']
        
        if type(city) != str or type(region) != str or type(country) != str:
            continue

        city_counts["%s, %s, %s" % (city, region, country)] += 1
    return city_counts
```

```python
city_counts = get_city_counts(funding)
```

```python
def get_region_labels(clusters, city_counts):
    city_to_region = {}
    for cluster in clusters:
        max_city = ""
        max_count = 0
        for i in cluster:
            if city_counts[i] > max_count:
                max_count = city_counts[i]
                max_city = i
        print max_city
        for j in cluster:
            city_to_region[j] = max_city
    return city_to_region
```

```python
city_to_region = get_region_labels(clusters, city_counts)
```

```python
dirty = open('crunchbase_data.csv', 'rb')
clean = open('clean_data.csv', 'wb')

reader = csv.reader(dirty)
writer = csv.writer(clean)

header = next(reader)
writer.writerow(['region', 'coordinates'] + header[3:])

for row in reader:
    country = row[0]
    region = row[1]
    city = row[2]
        
    if type(city) != str or type(region) != str or type(country) != str:
        continue

    lookup_string = "%s, %s, %s" % (city, region, country)
    if lookup_string not in city_to_region:
        continue
    region = city_to_region[lookup_string]
    if region not in x:
        continue
    writer.writerow([region] + [x[region]] + row[3:])
```
